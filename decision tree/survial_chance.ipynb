```python
import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\malia\OneDrive\Desktop\Machine learning\decision tree\Titanic-Dataset.csv')
print(df.info())
```

```python
cols=df[['Fare','Pclass','Sex','Age']]

from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
cols['Sex_n']=le_sex.fit_transform(cols['Sex'])
cols=cols.drop(['Sex'],axis='columns')
print(cols)
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
X=cols
y=df['Survived']
x_train , x_test , y_train, y_test=train_test_split(X,y,test_size=0.2)
dec_tr=tree.DecisionTreeClassifier()
model=dec_tr.fit(x_train,y_train)

print(model.score(x_test,y_test))
```
